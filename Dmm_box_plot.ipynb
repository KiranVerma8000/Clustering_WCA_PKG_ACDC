```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file containing normalized DMM metrics
excel_file = "./inputs/commit_metrics.xlsx"
data = pd.read_excel(excel_file)

desired_add_types = ["Existence", "Property", "Executive"]  # Adjust based on your analysis needs
design_decision_datasets = {add_type: {metric_type: [] for metric_type in ['DMM Unit Size', 'DMM Unit Complexity', 'DMM Unit Interfacing']} for add_type in desired_add_types}

for index, row in data.iterrows():
    add_types_str = row['ADD']  
    add_types = eval(add_types_str) 
    for add_type in desired_add_types:
        add_value = add_types[add_type].replace("'", "").replace('"', '')
        if add_value == 'True':  
            for metric_type in ['DMM Unit Size', 'DMM Unit Complexity', 'DMM Unit Interfacing']:
                design_decision_datasets[add_type][metric_type].append(row[metric_type])        
df_datasets = {}
for add_type in desired_add_types:
    df_datasets[add_type] = pd.DataFrame(design_decision_datasets[add_type])

fig, axes = plt.subplots(nrows=1, ncols=len(desired_add_types), figsize=(20, 7), sharey=True)
for idx, metric_type in enumerate(desired_add_types):
    sns.boxplot(data=df_datasets[metric_type], ax=axes[idx])
    axes[idx].set_title(f"Design Decision: {metric_type}")
    axes[idx].set_xlabel("DMM Metrics Type")
    axes[idx].set_ylabel("DMM Metrics Value")
    axes[idx].tick_params(axis='x', labelrotation=10) 

plt.suptitle("Box Plot of DMM Metrics by Design Decision", y=1.05)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from scipy.stats import kruskal
import scikit_posthocs as sp

# Load the Excel file containing normalized DMM metrics
excel_file = "./inputs/commit_metrics.xlsx"
data = pd.read_excel(excel_file)

desired_add_types = ["Existence", "Property", "Executive"]  # Adjust based on your analysis needs
design_decision_datasets = {add_type: {metric_type: [] for metric_type in ['DMM Unit Size', 'DMM Unit Complexity', 'DMM Unit Interfacing']} for add_type in desired_add_types}

for index, row in data.iterrows():
    add_types_str = row['ADD']  
    add_types = eval(add_types_str) 
    for add_type in desired_add_types:
        add_value = add_types[add_type].replace("'", "").replace('"', '')
        if add_value == 'True':  
            for metric_type in ['DMM Unit Size', 'DMM Unit Complexity', 'DMM Unit Interfacing']:
                design_decision_datasets[add_type][metric_type].append(row[metric_type])        

# Perform Kruskal-Wallis H-test
kruskal_results = {}
for metric_type in ['DMM Unit Size', 'DMM Unit Complexity', 'DMM Unit Interfacing']:
    print(f"\nKruskal-Wallis H-test for {metric_type}:")
    groups = [design_decision_datasets[add_type][metric_type] for add_type in desired_add_types]
    H, p_value = kruskal(*groups)
    print(f"H-statistic: {H}")
    print(f"P-value: {p_value}")
    kruskal_results[metric_type] = (H, p_value)

# Perform Dunn's post hoc test
for metric_type, (H, p_value) in kruskal_results.items():
    if p_value < 0.05:  # If Kruskal-Wallis test is significant
        print(f"\nDunn's post hoc test for {metric_type}:")
        data_for_dunn = [design_decision_datasets[add_type][metric_type] for add_type in desired_add_types]
        dunn_results = sp.posthoc_dunn(data_for_dunn, p_adjust='bonferroni')
        print(dunn_results)
    else:
        print(f"\nDunn's post hoc test for {metric_type} is not applicable as Kruskal-Wallis test is not significant.")
```
