```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
def prepare_data(json_path, commit_hashes):
    with open(json_path, 'r') as f:
        data = json.load(f)
    
    # Initialize a list to store the extracted data
    extracted_data = []
    
    # Iterate over each commit hash
    for key, value in data.items():
        for commit_hash in commit_hashes:
            if commit_hash in key:
                extracted_data.append(float(value))
            # else:
            #     # If the commit hash doesn't exist in the JSON data, append None
            #     extracted_data.append(None)
    
    
    return extracted_data
```

```python
acdc_json_path = 'A2A_metrics/a2a_metric_acdc.json'
# acdc_json_path = 'Normalized_data_A2A_metrics/Normalized_a2a_metric_acdc.json'
path = 'inputs/filtered_commit_metrics.xlsx'
column_name = 'Commit Hash'
commit_metrics_df_existance = pd.read_excel(path, sheet_name="Existence_True")[column_name].tolist()
commit_metrics_df_Property = pd.read_excel(path, sheet_name="Property_True")[column_name].tolist()
commit_metrics_df_Executive = pd.read_excel(path, sheet_name="Executive_True")[column_name].tolist()

data_existence = prepare_data(acdc_json_path, commit_metrics_df_existance)
data_property = prepare_data(acdc_json_path, commit_metrics_df_Property)
data_executive = prepare_data(acdc_json_path, commit_metrics_df_Executive)


max_length = max(len(data_existence), len(data_property), len(data_executive))

# Pad the shorter lists with None values to match the maximum length
data_existence += [None] * (max_length - len(data_existence))
data_property += [None] * (max_length - len(data_property))
data_executive += [None] * (max_length - len(data_executive))

# Combine the data into a DataFrame
df_data = pd.DataFrame({
    'Existence': data_existence,
    'Property': data_property,
    'Executive': data_executive
})

# Create a box plot using Seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_data)
plt.title('Box Plot of ACDC clustering coverage by Design Decision')
plt.xlabel('Design Decision')
plt.ylabel('ACDC clustering coverage')
plt.show()
```

```python
from scipy.stats import kruskal

# Example data (replace with your actual data)
group1_data = df_data['Existence'].dropna()
group2_data = df_data['Property'].dropna()
group3_data = df_data['Executive'].dropna()

# Perform Kruskal-Wallis H-test
H, p_value = kruskal(group1_data, group2_data, group3_data)

# Interpret the result
print("Kruskal-Wallis H-test:")
print(f"H-statistic: {H}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("The differences between groups are statistically significant.")
else:
    print("There is no significant difference between groups.")
```

```python
import scikit_posthocs as sp

# Example data (replace with your actual data)
group1_data = df_data['Existence'].dropna()
group2_data = df_data['Property'].dropna()
group3_data = df_data['Executive'].dropna()

# Perform Kruskal-Wallis H-test (if not already performed)
H, p_value = kruskal(group1_data, group2_data, group3_data)

# If the Kruskal-Wallis test is significant, proceed with Dunn's post hoc test
if p_value < 0.05:
    # Create a dictionary mapping group names to their data
    data = {
        'Existence': group1_data,
        'Property': group2_data,
        'Executive': group3_data
    }
    
    # Perform Dunn's test
    dunn_results = sp.posthoc_dunn(list(data.values()), p_adjust='bonferroni')
    
    # Print Dunn's test results
    print("Dunn's Test Results:")
    print(dunn_results)
else:
    print("No significant difference detected. Dunn's test is not applicable.")
```
